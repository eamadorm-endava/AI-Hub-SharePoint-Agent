## Requirements to login to Microsoft Graph API

### **Register an App in Microsoft Entra**

You first need to register an application on [Microsoft Entra](https://learn.microsoft.com/es-mx/graph/auth/auth-concepts). 

### **Important IDs related to the registered App**

Once you successfully registered the application, you will get:

- ***Application (customer) ID***: Uniquely identifies the application in the Microsoft cloud ecosystem, across all tenants.
- **Object ID**: Uniquely identifies the application in your tenant

### **Add Credentials**

Then you need to add **Credentials**, which are used by sensitive client applications that access a web API. **It allows the application to authenticate as itself, without requiring any user interaction at runtime**

You can add certificates, client secrets (a string or password), or federated credentials as credentials to the confidential client application registry

##### **Certificates**

Sometimes referred to as a *public key*, a certificate is the recommended type of credential because they are considered more secure than client secrets.


##### **Client Secret (this type of credential will be used in the project)**

Also referred to as an application password, a client secret is a string value that the application can use in place of a certificate for its own identity. Client secrets are considered less secure than certificate credentials. Sometimes, application developers use client secrets during local application development due to their ease of use. However, you must use certificate credentials or federated credentials for applications running in production.

```python
azure_config = AzureConfig()


azure_config.load_secrets()
```

Initialize the authentication provider

```python
credential = ClientSecretCredential(
    tenant_id = azure_config.AIHUB_APP_TENANT_ID,
    client_id = azure_config.AIHUB_APP_CLIENT_ID,
    client_secret=azure_config.AIHUB_APP_CLIENT_SECRET_VALUE.get_secret_value()
)
```

```python
scopes = ['https://graph.microsoft.com/.default']
```

```python
client = GraphServiceClient(credentials=credential, scopes=scopes)
```
